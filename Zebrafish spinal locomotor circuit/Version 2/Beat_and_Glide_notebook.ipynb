# Zebrafish Swimming Models
### Below is the block of code to run beat-and-glide swimming models

```python
### FIRST, run this code to set the simulation parameters. Then select from the codes below to run different models

%matplotlib inline


#Declare the duration of the simulation
TIME_END = 5000
SKIP = 1000
PLOT_INTERVAL = 5000
PLOT_RESULT = True
PLOT_ALL = False
PRINT_PARAM = False
SAVE_CSV = True
SAVE_ANIM = True
DT = 0.1

import winsound
duration = 1000  # milliseconds
freq = 440  # Hz
```

### Block to run the base model

```python
#Base Model
%matplotlib inline
from datetime import datetime

from Beat_and_glide import Beat_and_glide_base

print("Base Model")
test = Beat_and_glide_base(stim0 = 2.89, sigma = 0, sigma_LR = 0.1,
                 E_glu = 0, E_gly = -70, cv = 0.80, 
                 nMN=15, ndI6=15, nV0v=15, nV2a=15, nV1=15, nMuscle=15, 
               R_str=1.0)
test.setTimeParameters(tmax_ms = TIME_END, tskip_ms = SKIP, dt = 0.1)  
((VLMN, VRMN), (VLdI6, VRdI6), (VLV0v, VRV0v), (VLV2a, VRV2a),
   (VLV1, VRV1),(VLMuscle, VRMuscle), Time) = test.mainLoop(rand=datetime.now(), tplot_interval_ms = PLOT_INTERVAL, plotAllPotentials=PLOT_ALL,
                                                      plotResult = PLOT_RESULT, printParam = PRINT_PARAM, saveCSV = SAVE_CSV, saveAnim =  SAVE_ANIM)
                                                     
#The code below produces a beeping sound af the code has been executed
winsound.Beep(freq, duration) 
```

### Block to run the V1 null model

```python
#V1 Knockout
%matplotlib inline
from datetime import datetime

# Import connectome package
from Beat_and_glide_V1_KO import Beat_and_glide_V1_KO

print("V1 Knockout Model")
test = Beat_and_glide_V1_KO(stim0 = 2.89, sigma = 0, sigma_LR = 0.1,
                 E_glu = 0, E_gly = -70, cv = 0.80, 
                 nMN=15, ndI6=15, nV0v=15, nV2a=15, nV1=15, nMuscle=15, 
               R_str=1.0)

#the silencing of V1 neurons is set below by tV1KOstart_ms and tV1KOend_ms
test.setTimeParameters(tmax_ms = TIME_END, tskip_ms = SKIP, dt = 0.1, tV1KOstart_ms=5000, tV1KOend_ms=8000)  
((VLMN, VRMN), (VLdI6, VRdI6), (VLV0v, VRV0v), (VLV2a, VRV2a),
   (VLV1, VRV1),(VLMuscle, VRMuscle), Time) = test.mainLoop(rand=datetime.now(), tplot_interval_ms = PLOT_INTERVAL, plotAllPotentials=PLOT_ALL,
                                                      plotResult = PLOT_RESULT, printParam = PRINT_PARAM, saveCSV = SAVE_CSV, saveAnim =  SAVE_ANIM)
                                                     
#The code below produces a beeping sound as the code has been executed
winsound.Beep(freq, duration) 
```
