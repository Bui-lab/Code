# Zebrafish single coiling model

### Below is the block of code to run single coiling models

```python
#Run first to initialize stimulation parameters
%matplotlib inline

from Util import PlaySound

TIME_END = 10000
SKIP = 200
PLOT_INTERVAL = 50000
PLOT_RESULT = True
PRINT_PARAM = False
SAVE_CSV = True
SAVE_ANIM = True
```

### Block to run the base model (note that only the left side to avoid the confound of coils on both side that occur coincidentally rather than due to synaptic linkage)

```python
#Single coil base model
from  Single_coiling_model import *

print("Basic Model")
test = Single_coil_base(dt = 0.1, stim0 = 50, sigma = 0, 
                 E_glu = 0, E_gly = -45, cv = 4, nIC = 5, nMN = 10, nV0d = 10, nMuscle = 10)

test.mainLoop(rand=0, tmax = TIME_END, tskip = SKIP, tplot_interval = PLOT_INTERVAL, plotResult = PLOT_RESULT, printParam = PRINT_PARAM, saveCSV = SAVE_CSV, saveAnim =  SAVE_ANIM)

PlaySound()
```
